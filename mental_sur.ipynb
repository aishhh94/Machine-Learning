```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\ASHWARY MAN\OneDrive\Desktop\survey.csv")
df.head(5)
```

```python
df.drop(["Timestamp", "comments", "state"], axis=1, inplace=True)
df.head(5)
```

```python
df = df[(df['Age'] >= 18) & (df['Age'] <= 70)]
```

```python
df['Gender'] = df['Gender'].str.lower().str.strip()
```

```python
male_terms = [
    'male','m','man','cis male','cis man','mail','make','mal','male-ish',
    'maile','msle','male (cis)','malr'
]

df['Gender'] = df['Gender'].replace(male_terms, 'male')
```

```python
female_terms = [
    'female','f','woman','female ','cis female','cis-female/femme',
    'femake','femail','female (cis)'
]

df['Gender'] = df['Gender'].replace(female_terms, 'female')
```

```python
df.loc[~df['Gender'].isin(['male','female']), 'Gender'] = 'other'
```
