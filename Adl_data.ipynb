```python
cat_cols = ip.select_dtypes(include='object').columns
num_cols = ip.select_dtypes(exclude='object').columns
```

```python
cat_cols
```

```python
num_cols
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
ct = ColumnTransformer(transformers=[('cat', OneHotEncoder(), cat_cols),('num', 'passthrough', num_cols)])
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(ip, op, test_size=0.2, random_state=42)
```

```python
x_train = ct.fit_transform(x_train)
x_test = ct.transform(x_test)
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
pred = lr.predict(x_test)
pred
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
pred = lr.predict(x_test)
pred
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
ac = accuracy_score(pred, y_test)
re = recall_score(pred, y_test)
pr = precision_score(pred, y_test)
f1 = f1_score(pred, y_test)
```

```python
print('Accuracy Score: ',ac)
print('Recall Score: ',re)
print('Precision Score: ',pr)
print('F1 Score: ',f1)
```
